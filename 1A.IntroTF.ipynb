# Hello world of neural networks

```python
model = Sequential([ # successive layers denoted in sequence 
    Dense(units=1, input_shape=[1]) # fully connected neurons
])
```

```python
model.compile(optimizer='sgd', # takes in the information from the loss function to backpropogate
              loss='mean_squared_error')
```

```python
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
ys = np.array([-3, -1, 1, 3, 5, 7], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10])) # nearly 19 - very little data
```

# Assignment 

```python
def house_model(y_new):
    xs = np.array([1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([100, 150, 200, 250, 300, 350], dtype=float)
    model = Sequential([Dense(1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=50) # use more epochs
    predict = model.predict(y_new)[0]
    return predict*1000
```

```python
prediction = house_model([7.0])
print(prediction)
```
